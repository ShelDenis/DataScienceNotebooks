# Предобработка датасета "Стоимость поддержанного автомобиля"

```python
import pandas as pd

data = pd.read_csv('D:/PythonWork/OldZmiy/NLPWork/data/cars.csv')
data
```

```python
data = data.drop_duplicates()
data
```

```python
data.isna().sum()
```

```python
data = data.dropna()
data.isna().sum()
```

Преобразование нечисловых значений (на всякий случай для анализа сохраняем словарь соответствий)

```python
categories_mapping = {}
for col in data.select_dtypes(include=['object']).columns:
    unique_values = data[col].unique().tolist() 
    mapping_dict = {val: idx+1 for idx, val in enumerate(unique_values)}  
    categories_mapping[col] = mapping_dict
    
    data[col] = data[col].map(mapping_dict)

bool_cols = data.select_dtypes(include=['bool']).columns
data[bool_cols] = data[bool_cols].astype(int)

data
```

Удаление выбросов

```python
from matplotlib import pyplot as plt

cols = ['odometer_value', 'engine_capacity']

for c in cols:
    data.hist(c, bins=20, log=True)
    plt.show()
```

```python
outlier_fits = data[cols]

q1 = outlier_fits.quantile(0.25)
q3 = outlier_fits.quantile(0.75)

iqr = q3 - q1

data_filtered = outlier_fits[~((outlier_fits < (q1 - 1.5 * iqr)) | (outlier_fits > (q3 + 1.5 * iqr))).any(axis=1)]
```

```python
index_list = list(data_filtered.index.values)
data_filtered = data[data.index.isin(index_list)]

data_filtered.shape
```

```python
for c in cols:
    data_filtered.hist(c, bins=20, log=True)
    plt.show()
```

Масштабирование данных и понижение размерности

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
y = data_filtered['price_usd']
X = data_filtered.drop(['price_usd'], axis=1)

scaled_X = pd.DataFrame(scaler.fit_transform(X), columns=scaler.get_feature_names_out())
scaled_X
```

```python
from sklearn.tree import DecisionTreeRegressor
from matplotlib import pyplot as pyplot

tree = DecisionTreeRegressor().fit(scaled_X, y)

plt.barh(width=tree.feature_importances_, y=X.columns)
```

Оставим 7 признаков

```python
y
```

```python
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeRegressor

tree = DecisionTreeRegressor().fit(scaled_X, y)

rfe = RFE(estimator=tree, n_features_to_select=7, step=1).fit(scaled_X, y)
X_rfe = pd.DataFrame(rfe.transform(X), columns=rfe.get_feature_names_out(), index=X.index)
X_rfe
```

```python
short_df = pd.concat([X_rfe, y], axis=1)
short_df
```

```python
cols_to_int = ['manufacturer_name', 'model_name', 'odometer_value', 'year_produced', 'body_type', 'duration_listed']

short_df[cols_to_int] = short_df[cols_to_int].apply(lambda x: x.astype(int))
short_df.info()
```

```python
short_df.to_csv('cars_prepared.csv')
```
